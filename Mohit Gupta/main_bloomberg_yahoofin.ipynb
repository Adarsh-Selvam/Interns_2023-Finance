```python
import openpyxl
import matplotlib.pyplot as plt

# Load the workbook
wb = openpyxl.load_workbook(r'IBM Annual.xlsx')

# Select the active sheet
sheet = wb.active

# Read the values from the range L14:Z14
labelx = sheet["E14"].value
xaxis=[]
yaxis = []
for row in sheet['L14:Z14']:
    for cell in row:
        yaxis.append(cell.value)

for row in sheet['L11:Z11']:
    for cell in row:
        xaxis.append(cell.value)

plt.plot(xaxis,yaxis,label = labelx)
```
